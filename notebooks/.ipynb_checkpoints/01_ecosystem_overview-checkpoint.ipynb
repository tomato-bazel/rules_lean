# Lean ecosystem overview

Exploratory pass over `reservoir-manifest.json`. Goal: build a dependency graph and surface which packages are structurally interesting (load-bearing, hub, leaf, orphan, well-connected community).

Regenerate the manifest from repo root with: `python3 tools/reservoir_manifest.py`

```python
import json
from pathlib import Path

import networkx as nx
import pandas as pd

MANIFEST = Path('..') / 'reservoir-manifest.json'
raw = json.loads(MANIFEST.read_text())
print('upstreamBundledAt:', raw['upstreamBundledAt'])
print('fetchedAt        :', raw['fetchedAt'])
print('packageCount     :', raw['packageCount'])
print('toolchainCount   :', raw['toolchainCount'])
```

## Package table

Flatten each package to a single row of scalar/summary fields. The bulky nested arrays (`versions`, `builds`, `dependents`) stay in `raw` for graph construction.

```python
aliases = raw['packageAliases']  # old fullName -> current fullName

def canon(full_name: str) -> str:
    seen = set()
    while full_name in aliases and full_name not in seen:
        seen.add(full_name)
        full_name = aliases[full_name]
    return full_name

rows = []
for p in raw['packages']:
    src = (p.get('sources') or [{}])[0]
    latest_v = (p.get('versions') or [None])[-1] or {}
    rows.append({
        'fullName': p['fullName'],
        'owner': p['owner'],
        'name': p['name'],
        'stars': p['stars'],
        'license': p['license'],
        'createdAt': p['createdAt'],
        'updatedAt': p['updatedAt'],
        'host': src.get('host'),
        'repoUrl': src.get('repoUrl'),
        'versionCount': len(p.get('versions') or []),
        'buildCount': len(p.get('builds') or []),
        'dependentCount': len(p.get('dependents') or []),
        'latestToolchain': latest_v.get('toolchain'),
        'latestRev': latest_v.get('revision'),
    })
pkgs = pd.DataFrame(rows).set_index('fullName')
pkgs['createdAt'] = pd.to_datetime(pkgs['createdAt'], utc=True)
pkgs['updatedAt'] = pd.to_datetime(pkgs['updatedAt'], utc=True)
pkgs.shape, pkgs.head(3)
```

## Dependency graph

Edge convention: `A -> B` means *A depends on B*. The manifest stores the inverse (each package lists its `dependents`), so we flip while loading.

Nodes that show up only as dependents (i.e. not themselves indexed by Reservoir) are added as placeholder nodes flagged `indexed=False`.

```python
G = nx.DiGraph()

indexed = {canon(p['fullName']) for p in raw['packages']}
for p in raw['packages']:
    target = canon(p['fullName'])
    G.add_node(target, indexed=True, stars=p['stars'], owner=p['owner'])

for p in raw['packages']:
    target = canon(p['fullName'])
    for d in (p.get('dependents') or []):
        depender = canon(d['fullName'])
        if depender not in G:
            G.add_node(depender, indexed=False, stars=None, owner=d.get('scope'))
        G.add_edge(depender, target,
                   rev=d.get('rev'), inputRev=d.get('inputRev'),
                   transitive=d.get('transitive', False))

print(f'nodes: {G.number_of_nodes()}  (indexed={sum(1 for _,a in G.nodes(data=True) if a["indexed"])})')
print(f'edges: {G.number_of_edges()}')
print(f'density: {nx.density(G):.5f}')
print(f'is_dag: {nx.is_directed_acyclic_graph(G)}')
```

## Structural summary

WCCs (weakly connected components) tell us how fragmented the ecosystem is. Big components = highly interconnected. Cycles (if any) signal Lake `inputRev` divergence between packages and need closer inspection.

```python
wccs = sorted(nx.weakly_connected_components(G), key=len, reverse=True)
print(f'weakly connected components: {len(wccs)}')
print(f'largest WCC size            : {len(wccs[0])}')
print(f'top WCC sizes               : {[len(c) for c in wccs[:10]]}')

try:
    cycles = list(nx.simple_cycles(G))
    print(f'simple cycles found         : {len(cycles)}')
    if cycles:
        for c in cycles[:5]:
            print('  ', ' -> '.join(c + [c[0]]))
except Exception as e:
    print('cycle enumeration skipped:', e)
```
